```python
def fraction(l):
    print("L", l)
    log = np.log
    exp = np.exp
    
    v = l*log(l)
    print("V: ", v)
    d = l*v
    print("d: ", d)
    
    temp = (v - 32)*(v - 32)
    print('Exponent numerator: {}'.format(temp))
    print('Exponent Denominator: {}'.format(32*d))    
    exponent = -1*(temp/(32*d))
    print("Exponent: ", exponent)
    print(l*exp(exponent))
    print()
    return l*exp(exponent)
```
